```python
import pandas as pd
from pandasai import PandasAI
from pandasai.llm.openai import OpenAI
from dotenv import dotenv_values
```

```python
df = pd.read_csv("data/BenchmarkingSample20230607.csv")
df.head(20)
```

```python
config = dotenv_values(".env")
llm = OpenAI(api_token = config['OPENAI_API_KEY'])
```

```python
pandas_ai = PandasAI(llm, enforce_privacy=True)
pandas_ai(df, prompt="Do a time series decomposition of the data")
```
