```python
df = pd.read_csv('data/BenchmarkingSample20230607_with_features.csv')
df
```

# Mean Analysis

```python
# Calculate the SMA with a window size of 3
df['Total Cost of Sales_SMA'] = df['Total Cost of Sales'].rolling(window=3).mean()
df['Gross Profit_SMA'] = df['Gross Profit'].rolling(window=3).mean()
df['Total Labor_SMA'] = df['Total Labor'].rolling(window=3).mean()
df['Total Prime Costs_SMA'] = df['Total Prime Costs'].rolling(window=3).mean()
df['Total Controllable Costs_SMA'] = df['Total Controllable Costs'].rolling(window=3).mean()
df['Controllable Profit_SMA'] = df['Controllable Profit'].rolling(window=3).mean()
df['Total Non-controllable Costs_SMA'] = df['Total Non-controllable Costs'].rolling(window=3).mean()
df['Store EBITDA_SMA'] = df['Store EBITDA'].rolling(window=3).mean()


df = df.round(2)
df
```
